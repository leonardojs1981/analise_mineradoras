## Análise do financiamento eleitoral por empresas de mineração

#### Basicamente simples manipulações de dados no arquivo do TSE, baixado em http://www.tse.jus.br/eleicoes/estatisticas/repositorio-de-dados-eleitorais-1/repositorio-de-dados-eleitorais

```python
import pandas as pd
df_receitas_cand_2014 = pd.read_csv('dados/prestacao_final_2014/receitas_candidatos_2014_brasil.txt', sep=';', encoding = 'ISO-8859-1', low_memory=False)
```

```python
df_receitas_cand_2014['valor_corrigido'] = df_receitas_cand_2014['Valor receita'].apply(lambda x: float(x.replace(',','.')))
```

#### Pausa pra identificação das empresas do ramo de mineração

```python
lista_mineracao = [k for k in set(df_receitas_cand_2014['Setor econômico do doador']) if 'miner' in k]
lista_mineracao_2 = [k for k in set(df_receitas_cand_2014['Setor econômico do doador']) if 'minér' in k]
```

```python
lista_mineracao = set(lista_mineracao + lista_mineracao_2)
lista_mineracao
```

```python
lista_atividades_mineracao = ['Atividades de apoio à extração de minerais não-metálicos',\
 'Beneficiamento de minério de estanho',\
 'Comércio atacadista de máquinas, equipamentos para terraplenagem, mineração e construção; partes e peças',\
 'Comércio atacadista de produtos da extração mineral, exceto combustíveis',\
 'Extração de carvão mineral',\
 'Extração de minério de alumínio',\
 'Extração de minério de estanho',\
 'Extração de minério de ferro',\
 'Extração de minério de manganês',\
 'Extração de minério de metais preciosos',\
 'Extração de minérios de cobre, chumbo, zinco e outros minerais metálicos não-ferrosos não especificados anteriormente',\
 'Extração de outros minerais não-metálicos não especificados anteriormente',\
 'Fabricação de outros produtos de minerais não-metálicos não especificados anteriormente',\
 'Manutenção e reparação de máquinas e equipamentos para uso na extração mineral, exceto na extração de petróleo',\
 'Representantes comerciais e agentes do comércio de combustíveis, minerais, produtos siderúrgicos e químicos',\
 'Produção de ferroligas']

```

```python
def mineradoras (x):
    if x in lista_atividades_mineracao:
        y = 1
    else:
        y = 0
    return y
```

#### Mais faxina

```python
df_receitas_cand_2014['flag_mineradoras'] = df_receitas_cand_2014['Setor econômico do doador'].map(mineradoras)
df_receitas_cand_2014['flag_mineradoras_2'] = df_receitas_cand_2014['Setor econômico do doador originário'].map(mineradoras)
df_receitas_cand_2014['ind_mineradoras'] = df_receitas_cand_2014['flag_mineradoras'] + df_receitas_cand_2014['flag_mineradoras_2']
```

```python
df_doadoras_mineradoras = df_receitas_cand_2014[df_receitas_cand_2014['ind_mineradoras'] >= 1][['CPF/CNPJ do doador','Nome do doador (Receita Federal)','Setor econômico do doador','CPF/CNPJ do doador originário','Nome do doador originário (Receita Federal)','Setor econômico do doador originário','Data da receita','valor_corrigido','Tipo receita','UF','Sigla  Partido','Numero candidato','Cargo','Nome candidato','CPF do candidato']]

```

```python
df_doadoras_mineradoras
```

```python
df_doadoras_mineradoras_1 = df_doadoras_mineradoras[df_doadoras_mineradoras['CPF/CNPJ do doador originário'] == '#NULO']
df_doadoras_mineradoras_2 = df_doadoras_mineradoras[df_doadoras_mineradoras['CPF/CNPJ do doador originário'] != '#NULO']
```

```python
df_doadoras_mineradoras_1['cnpj'] = df_doadoras_mineradoras_1['CPF/CNPJ do doador']
df_doadoras_mineradoras_1['nome_empresa_doadora'] = df_doadoras_mineradoras_1['Nome do doador (Receita Federal)']
df_doadoras_mineradoras_1['setor'] = df_doadoras_mineradoras_1['Setor econômico do doador']

df_doadoras_mineradoras_2['cnpj'] = df_doadoras_mineradoras_2['CPF/CNPJ do doador originário']
df_doadoras_mineradoras_2['nome_empresa_doadora'] = df_doadoras_mineradoras_2['Nome do doador originário (Receita Federal)']
df_doadoras_mineradoras_2['setor'] = df_doadoras_mineradoras_2['Setor econômico do doador originário']
```

```python
df_doadoras_mineradoras_ = pd.concat([df_doadoras_mineradoras_1, df_doadoras_mineradoras_2])[['Data da receita','valor_corrigido','Tipo receita','UF','Sigla  Partido','Numero candidato','Cargo','Nome candidato','CPF do candidato', 'cnpj', 'nome_empresa_doadora', 'setor']]
```
